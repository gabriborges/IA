### Interfaces do Matplotlib

O **estilo MATLAB** usa o pyplot (plt) como interface. Esta interface é `stateful` pois lembra das propriedades da figura atual. Bom para gráficos simples pois é mais direto e rápido.

A função **`plt.gcf()`** retorna a figura atual e a função **`plt.gca()`** retorna o Axe atual.

O estilo **orientado a objetos** acessa métodos diretamente nos objetos **Figure** e **Axe** que você precisar, e não através de pyplot.

As funções do estilo MATLAB são de fato delegações de chamadas para funções de objetos. A comunidade python recomenda a utilização orientada a objetos. Nesta aula vamos utilizar os dois estilo, dependendo da complexidade do exemplo. 

### Figures, Axes, Plots e Subplots

Plots em matplotlib são desenhados em um objeto `Figure`. Um `Figure` possui um ou vários `Axes`

```python
fig = plt.figure(figsize=(4,3))                 # cria uma objeto figura
ax = plt.axes()                    # cria um objeto axes
x = np.linspace(0,np.pi*4,50)
ax.plot(x, np.sin(x))               # cria o plot em um Axes
plt.show()
```

Alternativamente, você pode usar o estilo MATLAB e deixar a criação dos objetos por conta do matplotlib

```python
plt.plot(x, np.sin(x))
plt.show()
```

**Subplots** podem ser criados com a chamada a plt.subplot() em pyplot

```python
x = np.linspace(0,np.pi*4,100)
print(x)
plt.subplot(3, 1, 1)           # cria um painel (rows, columns, panel number)
plt.plot(x, np.sin(x))         # chama plot através do pyplot
plt.subplot(3, 1, 2)           # cria um segundo painel and set current axis
plt.plot(x, np.cos(x))         # chama plot através do pyplot
plt.subplot(3, 1, 3)
plt.plot(x, x)
```

A versão orientada a objetos usa a chamada plt.subplots para criar Figure e os Axes. Note que a chamada é diferente. Agora é **`plt.subplots`**

```python
fig, ax = plt.subplots(nrows=3) # cria um grid de plots
fig.axes[0].plot(x, np.sin(x))  # chama plot em cada objeto Axis
fig.axes[1].plot(x, np.cos(x))  # chama plot em cada objeto Axis
fig.axes[2].plot(x, x)  # chama plot em cada objeto Axis
```

Os objetos Axes podem ser desempacotados na chamada, assim não precisa usar `fig.axes`

```python
fig, (ax1,ax2) = plt.subplots(nrows=2, figsize=(8, 4)) # cria um grid de plots
```

```python
fig.axes[0] is ax1, fig.axes[1] is ax2 # duas alternativas para acessar o objeto Axis desejado
#fig.axes
```

```python
fig, axes = plt.subplots(nrows=2,ncols=3,figsize=(10,8))
axes[0, 0].plot([1, 2, 3, 4], [1, 4, 2, 3])
axes[0, 0].set_title('Eixo 0,0')

axes[0, 1].plot([1, 2, 3, 4], [4, 3, 2, 1])
axes[0, 1].set_title('Eixo 0,1')

axes[0, 2].plot([1, 2, 3, 4], [2, 2, 2, 2])
axes[0, 2].set_title('Eixo 0,2')

axes[1, 0].plot([1, 2, 3, 4], [3, 3, 3, 3])
axes[1, 0].set_title('Eixo 1,0')

axes[1, 1].plot([1, 2, 3, 4], [1, 2, 1, 2])
axes[1, 1].set_title('Eixo 1,1')

axes[1, 2].plot([1, 2, 3, 4], [2, 4, 2, 4])
axes[1, 2].set_title('Eixo 1,2')
```

#### ShareX e ShareY

```python
fig, axes = plt.subplots(nrows=2,ncols=3,sharex=True,sharey=True,figsize=(10,4))
#Quando você cria subplots com a opção sharex=True e sharey=True, você está especificando que todos os eixos compartilharão 
# o mesmo eixo x e o mesmo eixo y, respectivamente. Isso pode ser útil quando você deseja que todas as subplots tenham os 
# mesmos limites de eixos, escalas e rótulos.
```

#### Título, rótulos, ticks, legendas e anotações

Estilo matlab

```python
x = np.linspace(0,np.pi*2,25)
plt.title('Gráfico de 2*PI')
plt.xlabel('x')
plt.ylabel('sen(x)')
plt.plot(x, np.sin(x))
```

Estilo OO

```python
x = np.linspace(0,np.pi*2,25)
fig= plt.figure()
ax = plt.axes()
ax.set_title('Titulo do Plot')                  # propriedades usam getters e setters no modo OO
ax.set_xlabel('x')
ax.set_ylabel('sen(x)')
ax.plot(x, np.sin(x))
```

#### Cores, marcas e estilos de linhas

```python
x = np.linspace(0,np.pi*2, 25)
plt.title('Exemplo com cor e estilo de linhas')
plt.plot(x, np.sin(x), linestyle='--', color='red')
```

```python
plt.title('Exemplo com string de formatação')
plt.plot(x,np.sin(x), 'g.-') # forma alternativa
```

Marcas podem ser utilizadas para cada ponto de dado

```python
plt.title('Exemplo com marcador')
plt.plot(x,np.sin(x), 'go-',label='seno')
```

Para adicionar uma legenda nomeie antes cada serie de dados com o parâmetros `label`

```python
plt.title("Exemplo com legenda")
plt.plot(x,np.sin(x),label='seno')
plt.plot(x,np.cos(x),label='cosseno')
plt.legend(loc='best')
```

O método **ax.set()** permite ajustar todas as propriedades de uma vez

```python
x = np.linspace(0,np.pi*2,25)
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, np.pi*2), 
       ylim=(-1.25, 1.25),
       xlabel='x', 
       ylabel='sin(x)',
       title='Usando ax.set()');
```

Se você precisar de **multiplos plots** em uma mesma figura, apenas chame `plot()` várias vezes no mesmo comando

```python
plt.title("Multiplos Plots")
plt.plot(x,np.sin(x),label='seno')
plt.plot(x,np.cos(x),label='cosseno')
plt.legend()
```

### Histogramas

```python
x = np.random.randn(10000)
```

```python
print(x)
```

```python
plt.hist(x)
```

```python
plt.hist(x, bins=50, alpha=1, color='steelblue', edgecolor='black');
```

```python
x1 = np.random.normal(-5, 1.5, 1000)
x2 = np.random.normal(0, 3, 1000)
x3 = np.random.normal(6, 3, 1000)
params = dict(histtype='stepfilled', alpha=0.5, bins=30)
plt.hist(x1, **params)
plt.hist(x2, **params)
plt.hist(x3, **params);
```

Para gerar apenas contagem de itens use `np.histogram()`

```python
counts, bin_edges = np.histogram(x, bins=10)
print(counts)
list(zip(counts,bin_edges))
```

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x,y=np.random.multivariate_normal(mean, cov,10000).T
print(y)
```

```python
plt.scatter(x,y)
```

```python
plt.hist2d(x, y, bins=25, cmap='jet');
cb = plt.colorbar()
cb.set_label('Contagem em cada bin')
```

```python
plt.hexbin(x, y, gridsize=25, cmap='jet')
cb = plt.colorbar(label='count in bin')
```

```python
counts,xedges,yedges= np.histogram2d(x,y,bins=25)
counts
```

### Barras de Erro

```python
x = np.arange(10)
ruido = np.random.normal(8,3,1000).reshape(-1,10)
erros = ruido.var(axis=0)
y = 1.3*x + erros
```

```python
plt.hist(ruido.flatten(),bins=20);
```

```python
plt.errorbar(x,y,yerr=erros,fmt='o',ecolor='red')
```

### Plot com Pandas

Series e DataFrame possuem o metodo `plot()`

```python
import pandas as pd
```

```python
s = pd.Series(np.random.randn(100).cumsum())
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(100, 5).cumsum(0),
                  columns=['A', 'B', 'C', 'D','E'])
df.plot?
```

As opções do matplotlib podem ser passadas como parâmetro para o método `plot()`: figsize, subplots, sharex, sharey, title, legend, label, xlim, ylim, etc.

```python
df.plot(kind='hist',
        bins=50,
        alpha=0.5,
        edgecolor='k',
        figsize=(8,3),
        title="Histograma"
       );
```

O objeto DataFrame possui um atributo atributo `plot` com referencia para vários métodos de plotagem: line, bar, scatter, etc.

```python
# equivalente a df.plot()
df.plot.line() 
```

Pode-se usar o nome de colunas em um gráfico de dispersão 

```python
df.plot.scatter('B','E',alpha=0.5)
```
