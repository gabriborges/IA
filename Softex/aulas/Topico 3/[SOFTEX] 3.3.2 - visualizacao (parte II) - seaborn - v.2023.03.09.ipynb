## Estilos

````python
def sinplot(n=10, flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, n + 1):
        plt.plot(x, np.sin(x + i * .5) * (n + 2 - i) * flip)
````

````python
f = plt.figure(figsize=(8, 5))
gs = f.add_gridspec(2, 2)

with sns.axes_style("dark"):
    ax = f.add_subplot(gs[0, 0])
    ax.set_title("dark")
    sinplot(6)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[0, 1])
    ax.set_title("darkgrid")
    sinplot(6)

with sns.axes_style("whitegrid"):
    ax = f.add_subplot(gs[1, 0])
    ax.set_title("whitegrid")
    sinplot(6)

with sns.axes_style("ticks"):
    ax = f.add_subplot(gs[1, 1])
    ax.set_title("ticks")
    sinplot(6)

f.tight_layout()
````

#### Histogram and Distribution

````python
sns.set_palette('muted')
````

````python
sns.histplot(df.Attack)
````

````python
sns.histplot(tips.tip)
````

````python
sns.histplot(tips.tip,kde=True)
````

````python
fig,axes = plt.subplots(ncols=2,figsize=(8,3))
sns.histplot(tips.tip,ax=axes[0])
sns.kdeplot(tips.tip,ax=axes[1])
plt.tight_layout()
````

````python
sns.kdeplot(df.Attack,fill=True)
sns.kdeplot(df.Defense,fill=True)
````

````python
sns.kdeplot(data=df,x='Attack',y='Defense')
````

````python
# Distribution Plot (a.k.a. Histogram)
sns.displot(df.Attack)
````

````python
sns.jointplot(data=df,x="Attack",y= "Defense");
````

````python
sns.jointplot(data=df,x="Attack",y= "Defense",kind='kde');
````

````python
sns.jointplot(x="Attack",y= "Defense", data=df,kind='kde',fill=True);
````

Se a coluna usada no `hue` é numérica, precisamos usar o parametro `vars` para dizer ao seaborn quais são as colunas que desejamos usar no `pairplot` 

````python
sns.pairplot(data=df,
             vars=['HP','Attack','Defense','Speed'],
             hue='Stage',
             palette='bright')
````
